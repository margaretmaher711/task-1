```python
import networkx as nx
G=nx.Graph()
G.add_nodes_from(['a','b','c','d'])
G.add_edges_from([('a','b'),('a','c'),('b','c'),('c','d')])
nx.draw(G,with_labels=True,node_color="red",node_size=1000,font_size=20,font_color="black")
```

```python
dict(G.degree())
```

```python
nx.to_numpy_matrix(G)
```

```python
import pandas as pd
df =pd.read_csv('HomoSapiens_binary.txt',nrows=100,sep='\t',usecols=['Gene_A','Gene_B'])
df
```

```python
bio_graph=nx.from_pandas_edgelist(df,'Gene_A','Gene_B',create_using=nx.DiGraph)
nx.draw(bio_graph,with_labels=True,node_color="blue",node_size=1000,font_size=20,font_color="black")
```

```python
dd =pd.read_csv('SaccharomycesCerevisiaeS288C_binary_hq.txt',nrows=100,sep='\t',usecols=['Uniprot_A','Uniprot_B'])
bio_graph1=nx.from_pandas_edgelist(dd,'Uniprot_A','Uniprot_B',create_using=nx.DiGraph)
nx.draw(bio_graph1,with_labels=True,node_color="blue",node_size=1000,font_size=20,font_color="black")
```
